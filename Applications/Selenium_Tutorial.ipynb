# Testing that installation went well

```python
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
options = Options()
options.add_argument("--headless")
assert options.headless

driver = Chrome(curr_path,options = options)
driver.get('https://duckduckgo.com')

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

elem = driver.find_element(By.NAME, "q")
elem.clear()
elem.send_keys("pycon")
elem.send_keys(Keys.RETURN)
assert "No results found." not in driver.page_source
driver.close()
```

# Creating a non-headerless browser and clicking a button

```python
import time
sleep_time = 2

url ="https://pipeline2.kindermorgan.com/LocationDataDownload/LocDataDwnld.aspx?TSP=CPD"
button_id = "WebSplitter1_tmpl1_ContentPlaceHolder1_HeaderBTN1_btnRetrieve"
button_download_id = "WebSplitter1_tmpl1_ContentPlaceHolder1_HeaderBTN1_btnDownload"

driver = Chrome(curr_path)
driver.get(url)


elem = driver.find_element(By.ID,button_id)
elem.click()

time.sleep(sleep_time)

elem = driver.find_element(By.ID,button_download_id)
elem.click()

time.sleep(sleep_time)
```
